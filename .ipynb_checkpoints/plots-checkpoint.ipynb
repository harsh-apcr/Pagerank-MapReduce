```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
runtime1 = [17.15, 25.77, 35.74, 43.95, 55.27, 63.78, 72.75, 84.44, 96.18] 
runtime2 = [3.89, 4.73, 8.00, 11.32, 12.56, 14.54, 15.53, 17.86, 21.23]
runtime3 = [13.83, 20.48, 18.96, 19.90, 22.00, 24.29, 26.43, 31.17, 30.79]
size = [20000, 30000, 40000, 50000, 60000, 70000, 80000, 90000, 100000]
plt.figure(figsize=(10, 7))
plt.plot(size, runtime1, label='cpp')
plt.plot(size, runtime2, label='mpi')
plt.plot(size, runtime3, label='mpi-base')


plt.xlabel('benchmark-id')
plt.ylabel('mapreduce-pgrank-runtime (ms)')
plt.title('barabasi')
plt.legend()
```
